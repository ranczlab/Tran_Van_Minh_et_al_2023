```python
import numpy as np
import os
from pathlib import Path
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("ticks")
hfont = {'fontname':'Myriad Pro'}

import plotly.express as px
```

```python
path = Path(os.getcwd()).parents[0]
datapath = os.path.join(path, 'model_selection')
respath = os.path.join(datapath, 'residuals_distribution')
```

```python
ectpath = os.path.join(path, 'model_selection_external')
```

```python
AICc_res = pd.read_csv(os.path.join(respath, 'AICc_residual_distribution_all_targets.csv'))
```

```python
AICc_res =AICc_res.set_index('area')
```

```python
AICc_res2 = AICc_res.iloc[:-4]
```

```python
AICc_res2.index
```

```python
#AICc_res2 = AICc_res2.drop(AICc_res2.index[[18,19,21]])
```

```python
AICc_res2 = AICc_res2.sort_values(by = 'deltaAICc')
```

```python
# AICc_res_sub =AICc_res_sub.set_index('area')
```

```python
plt.figure(figsize = (13,10))
sns.heatmap(AICc_res2[['deltaAICc']],annot=True, square=True, cmap = 'vlag', vmin = AICc_res2.deltaAICc.min(), vmax = -AICc_res2.deltaAICc.min())
plt.savefig(os.path.join(datapath, 'heatmap_AICc_res_perarea.png'))
```
