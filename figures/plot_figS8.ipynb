```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
from matplotlib import gridspec
import seaborn as sns
sns.set_style("ticks")
import os
from pathlib import Path
import itertools
import sys
import pickle
import sys
sys.path.append(os.path.join(Path(os.getcwd()).parents[0], "functions"))
import warnings
warnings.filterwarnings('ignore')
import helper_functions as hfn
```

```python
path = Path(os.getcwd()).parents[0]

modelpath = os.path.join(path, 'data/network_models/probabilistic_model')
respath = os.path.join(modelpath, 'grid_simulations')

figpath = os.path.join(path, 'figures')
hfn.dir_check(figpath)

figs24path = os.path.join(figpath, 'FigS8')
hfn.dir_check(figs24path)

figs24panelpaths = os.path.join(figs24path, 'panels')
hfn.dir_check(figs24panelpaths)
```

```python
alldeg = pd.read_hdf(os.path.join(respath, 'alldeg.h5'))
```

```python
colors5000 = sns.color_palette("Purples_r", 10)
colors10000 = sns.color_palette("PuRd_r", 10)
colors50000 = sns.color_palette("Reds_r", 10)
colors200000 = sns.color_palette("YlOrBr_r", 10)
```

```python
flierprops={ 'markersize': 4,'markerfacecolor': 'None' }
boxprops = {'alpha':0.6}
```

```python
f = plt.figure(constrained_layout = True)
gs = f.add_gridspec(5,8)

ax_title1 = f.add_subplot(gs[0:1, :2])
plt.text(0.25, 0.5,'Ni = 5000', fontsize = 18)
ax_title1.set_axis_off()
ax_box1 = f.add_subplot(gs[1:2, :2])
ax_hist1 = f.add_subplot(gs[2:5, :2])


# f, (ax_box1, ax_hist1) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
sns.boxplot(data = alldeg[(alldeg.Ni == 5000)], 
            x = 'sdeg', 
            y= 'p', 
            orient="h", 
            ax=ax_box1, 
            palette = colors5000[0:5], 
            linewidth = 0.75, 
            flierprops=flierprops, 
            boxprops = boxprops)
h1 = sns.histplot(data = alldeg[(alldeg.Ni == 5000)], x = 'sdeg', hue = 'p', element="step", ax = ax_hist1, palette = colors5000[0:5], legend = False)

ax_box1.set_axis_off()
ax_hist1.spines['right'].set_visible(False)
ax_hist1.spines['top'].set_visible(False)
h1.set_xlabel("starters average degree",fontsize=16)
h1.set_ylabel("counts",fontsize=16)
h1.tick_params(labelsize=12)
# plt.xlim(-0.35, 1.75)
plt.locator_params(axis='x', nbins=4)


ax_title2 = f.add_subplot(gs[0:1, 2:4])
plt.text(0.25, 0.5,'Ni = 10000', fontsize = 18)
ax_title2.set_axis_off()
ax_box2 = f.add_subplot(gs[1:2, 2:4])
ax_hist2 = f.add_subplot(gs[2:5, 2:4])


sns.boxplot(data = alldeg[(alldeg.Ni == 10000)], 
            x = 'sdeg', 
            y= 'p', 
            orient="h", 
            ax=ax_box2, 
            palette = colors10000[0:6], 
            linewidth = 0.75, 
            flierprops=flierprops, 
            boxprops = boxprops)
h2 = sns.histplot(data = alldeg[(alldeg.Ni == 10000)], x = 'sdeg', hue = 'p', element="step", ax = ax_hist2, palette = colors10000[0:6], legend = False)

ax_box2.set_axis_off()
ax_hist2.spines['right'].set_visible(False)
ax_hist2.spines['top'].set_visible(False)
h2.set_xlabel("starters average degree",fontsize=16)
h2.set_ylabel("counts",fontsize=16)
h2.tick_params(labelsize=12)
# plt.xlim(-0.35, 1.75)
plt.locator_params(axis='x', nbins=4)


ax_title3 = f.add_subplot(gs[0:1, 4:6])
plt.text(0.25, 0.5,'Ni = 50000', fontsize = 18)
ax_title3.set_axis_off()
ax_box3 = f.add_subplot(gs[1:2, 4:6])
ax_hist3 = f.add_subplot(gs[2:5, 4:6])

sns.boxplot(data = alldeg[(alldeg.Ni == 50000)], 
            x = 'sdeg', 
            y= 'p', 
            orient="h", 
            ax=ax_box3, 
            palette = colors50000[0:6], 
            linewidth = 0.75, 
            flierprops=flierprops, 
            boxprops = boxprops)
h3 = sns.histplot(data = alldeg[(alldeg.Ni == 50000)], x = 'sdeg', hue = 'p', element="step", ax = ax_hist3, palette = colors50000[0:6], legend = False)

ax_box3.set_axis_off()
ax_hist3.spines['right'].set_visible(False)
ax_hist3.spines['top'].set_visible(False)
h3.set_xlabel("starters average degree",fontsize=16)
h3.set_ylabel("counts",fontsize=16)
h3.tick_params(labelsize=12)
# plt.xlim(-0.35, 1.75)
plt.locator_params(axis='x', nbins=4)

ax_title4 = f.add_subplot(gs[0:1, 6:])
plt.text(0.25, 0.5,'Ni = 200000', fontsize = 18)
ax_title4.set_axis_off()
ax_box4 = f.add_subplot(gs[1:2, 6:])
ax_hist4 = f.add_subplot(gs[2:5, 6:])

sns.boxplot(data = alldeg[(alldeg.Ni == 200000)],
            x = 'sdeg', 
            y= 'p', 
            orient="h", 
            ax=ax_box4, 
            palette = colors200000[0:6], 
            linewidth = 0.75, 
            flierprops=flierprops, 
            boxprops = boxprops)
h4 = sns.histplot(data = alldeg[(alldeg.Ni == 200000)], x = 'sdeg', hue = 'p', element="step", ax = ax_hist4, palette = colors200000[0:6], legend = False)

ax_box4.set_axis_off()
ax_hist4.spines['right'].set_visible(False)
ax_hist4.spines['top'].set_visible(False)
h4.set_xlabel("starters average degree",fontsize=16)
h4.set_ylabel("counts",fontsize=16)
h4.tick_params(labelsize=12)
# plt.xlim(-0.35, 1.75)
plt.locator_params(axis='x', nbins=4)

ax_box1.set_rasterized(True)
ax_box2.set_rasterized(True)
ax_box3.set_rasterized(True)
ax_box4.set_rasterized(True)
ax_hist1.set_rasterized(True)
ax_hist2.set_rasterized(True)
ax_hist3.set_rasterized(True)
ax_hist4.set_rasterized(True)
plt.gcf().set_size_inches(16,5)

plt.savefig(os.path.join(figs24panelpaths, 'figS8a.png'), bbox_inches = 'tight')
plt.savefig(os.path.join(figs24panelpaths, 'figS8a.eps'), bbox_inches = 'tight', format = 'eps', dpi = 200)
```
