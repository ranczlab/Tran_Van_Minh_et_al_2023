# Make lists to filter data

```python
# define list of areas (raw dataset)
list_areas_raw = list(df_raw_ipsi.columns)[11:]
```

```python
list_VIS = ['VISp', 'VISpm', 'VISl', 'VISam', 'VISal']

list_Thal = ['AM', 'LD', 'LP', 'LGd']

list_Thal_noLGd = ['AM', 'LD', 'LP']

list_RSP = ['RSPagl', 'RSPd', 'RSPv']

list_Dist_ctx = ['ORB', 'ACA', 'MOs', 'CLA', 'PTLp', 'TEa', 'AUD']

list_pooled = ['VIS', 'RSP', 'Thal', 'Thal_noLGd', 'Dist_ctx']

list_final_unpooled = list_VIS + list_RSP + list_Thal + list_Dist_ctx
list_final = list_VIS + list_RSP + list_Thal + list_Dist_ctx + list_pooled
```

# Filter data

```python
# apply mice subselection to raw datasets
df_raw_ipsi = df_raw_ipsi[df_raw_ipsi['name'].isin(list_mice_selected)].reset_index(drop=True)
```

```python
df_ipsi = df_raw_ipsi.copy()
```

```python
# add pooled areas to df_ipsi
df_ipsi['Thal'] = df_raw_ipsi[list_Thal].sum(axis = 1)
df_ipsi['Thal_noLGd'] = df_raw_ipsi[list_Thal_noLGd].sum(axis = 1)
df_ipsi['Dist_ctx'] = df_raw_ipsi[list_Dist_ctx].sum(axis = 1)
```

```python
list_final
```

```python
df_ipsi.rename(columns = {'Ratio of V1orPM from all startes':'ratio_target_all'}, inplace = True) 
```

```python
df_ipsi = df_ipsi.drop(columns = {'V1/PM'})
```

```python
# filter columns, make dataframe with all sub areas

df_filt_ipsi = df_ipsi[['name', 'pre', 'input', 'target area', 'strain', 'group', 'starter',
       'ratio_target_all', 'area', 'keep', 'VISp', 'VISpm', 'VISl',
       'VISam', 'VISal', 'RSPagl', 'RSPd', 'RSPv', 'AM', 'LD', 'LP', 'LGd',
       'ORB', 'ACA', 'MOs', 'CLA', 'PTLp', 'TEa', 'VIS', 'AUD',
       'RSP', 'Thal', 'Thal_noLGd', 'Dist_ctx']]
```

```python
# get extra metadata
col_subset = ['name', 'target area', 'strain', 'group','starter', 'ratio_target_all', 'area', 'keep']
```

```python
df_sub = df_ipsi[col_subset].set_index('name')
```

```python
mouse_dict = df_sub.to_dict('dict')

for k in mouse_dict.keys():
    df_filt_ipsi[k] = df_ipsi['name'].map(mouse_dict[k])
    
```

```python
# rename one column
df_filt_ipsi.rename(columns = {'Ratio of V1orPM from all startes':'ratio_target_all'}, inplace = True) 
```

```python
cols_ipsi = list(df_filt_ipsi.columns)

# reorder columns 
cols_ipsi_new = cols_ipsi[0:10] + cols_ipsi[-8:] +  cols_ipsi[10:-8]
df_filt_ipsi = df_filt_ipsi[cols_ipsi_new]
```

```python
df_pooled_ipsi = df_filt_ipsi[['name', 'pre', 'input', 'target area', 'strain', 'group','starter', 'ratio_target_all', 'area', 'keep'] + list_pooled]
```

```python
df_unpooled_ipsi = df_filt_ipsi[['name', 'pre', 'input', 'target area', 'strain', 'group','starter', 'ratio_target_all', 'area', 'keep'] + list_final_unpooled]
```
